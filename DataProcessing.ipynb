```python
regex = r'^(\d{4})'
extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype
```

```python
df['Date of Publication'].isnull().sum() / len(df)

# get mean of values in column S2
mean_value=df['Date of Publication'].mean()
```

```python
df['Date of Publication'].fillna(value=df['Date of Publication'].mean(), inplace=True)
#df['Date of Publication'].round(decimals=1)
df['Date of Publication'] = df['Date of Publication'].apply(np.ceil)
df
```

```python
university_towns = []
with open('/content/drive/MyDrive/DataVisualization/university_towns.txt') as file:
  for line in file:
    if '[edit]' in line:
        # Remember this `state` until the next is found
        state = line
    else:
        # Otherwise, we have a city; keep `state` as last-seen
        university_towns.append((state, line))

university_towns[:5]
```

```python
towns_df = pd.DataFrame(university_towns, columns=['State', 'RegionName'])
towns_df
```

```python
def get_citystate(item):
  if ' (' in item:
      return item[:item.find(' (')]
  elif '[' in item:
      return item[:item.find('[')]
  else:
      return item
```

```python
towns_df =  towns_df.applymap(get_citystate)
towns_df
```
